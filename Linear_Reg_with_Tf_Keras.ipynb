```
import numpy as np
import tensorflow as tf
from tensorflow import keras
```

```
x_batch = np.linspace(0, 1, 1000)
y_batch = 1.5 * x_batch + np.random.randn(*x_batch.shape) * 0.2 + 0.5
print (x_batch.shape)
```

```
regressor = keras.Sequential([
    keras.layers.Dense(64, activation=tf.nn.relu, input_shape=[1]),
    keras.layers.Dense(64, activation=tf.nn.relu),
    keras.layers.Dense(1)
  ])
 
optimizer = tf.keras.optimizers.Adam(0.001)
```

```
regressor.compile(loss='mean_squared_error',
                optimizer=optimizer,
                metrics=['mean_absolute_error', 'mean_squared_error'])
```

```
regressor.fit(x_batch,y_batch,epochs=500)
```

```
x = 65256
y = 1.5*x +0.5 #+ random_noise
print (y)
print (regressor.predict([x])[0][0])
```

```
splnn = keras.Sequential([   keras.layers.Dense(1, input_shape=[1]) ]) 
optimizer = tf.keras.optimizers.Adam(0.001)
```

```
splnn.compile(loss='mean_squared_error',
                optimizer=optimizer,
                metrics=['mean_absolute_error', 'mean_squared_error'])
```

```
splnn.fit(x_batch,y_batch,epochs=500)
```

```
x = 5896
y = 1.5*x +0.5
print (f'Actual: \t\t{y}')
print (f'Regressor: \t {regressor.predict([x])[0][0]}')
print (f'SPLNN: \t\t {splnn.predict([x])[0][0]}')
```
